# 3. How do changes of global carbon emission payback to European government?


In this question, we are going to investigate the effect of carbon emissions from three perspectives, 
- the effect on air quality
- the effect on agricultural production
- the effect on human health 

After looking for and researching the relative data, we take the following data as indicators of the above interested perspectives,
- tiniest pollution particles--PM2.5 pollution
- the FAO indices of agricultural production
- number of premature deaths and number of years of life lost due to PM2.5

A SQL database created in this section, named as ceproject.db, since the amount of data we have is large and data is miscellaneous from different indicators.

The dataset we found are mostly for countries all over the world. Therefore, a procedure is mostly appied for importing our data to our SQL database:
1. Import data, form it into a pandas Dataframe.
2. Select only the variables that could be relevant for the purposes of this analysis.
3. Drop duplicates and missing data, seperate the big Dataframe to smaller and more precise Dataframes for later convenience in SQL and analysis.
4. Filter out the countries that are in our consideration and convert them into nested lists form.
5. Transform the nested lists into tables in SQL database by sqlite3 commands.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import numpy as np
import sqlite3
import json
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

## 1.1 Country lists in Europe

```python
url = "https://en.wikipedia.org/wiki/List_of_European_countries_by_area"
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
table = soup.find('table', class_='wikitable sortable')
rows = table.find_all('tr')
countries_name = []
for row in rows:
    cells = row.find_all('td')
    if len(cells) > 0:
        country = cells[1].text.replace("*", "").strip()
        countries_name.append(country)
countries_name.remove('Vatican City')
print(countries_name)
```

## 1.2 Algricultural Production Index

The FAO indices of agricultural production represents the relative level of the aggregate volume of agricultural production for each year in comparison with the base period 2014-2016. The data is collected by Food and Agriculture Organization of the United Nations (FAO). 

```python
countries_df = pd.DataFrame(countries_name)
rawdata = pd.read_csv('data/Production_Indices_E_All_Data.csv',  encoding = 'ISO-8859-1')
rawdata = rawdata[(rawdata['Element'] == 'Gross Production Index Number (2014-2016 = 100)') & (rawdata['Item'] == 'Agriculture')]
header1 = ['Area']
header2 = ['Countries']
for x in range(1961, 2021, 1):
    x = str(x)
    header2.append(x)
    x = 'Y'+x
    header1.append(x)
rawdata = rawdata[header1]
rawdata = rawdata.set_axis(header2, axis=1)
rawdata = rawdata.replace('United Kingdom of Great Britain and Northern Ireland','United Kingdom')
rawdata = rawdata.dropna().reset_index(drop=True)
newdata = []
for c in countries_name:
    for a in range(len(rawdata)):
        if c in rawdata['Countries'][a]:
            for y in range(1961, 2021, 1):
                value = rawdata[str(y)][a]
                if np.isnan(value) == False:
                    temp = []
                    temp.append(c)
                    temp.append(y)
                    temp.append(rawdata[str(y)][a])
                    newdata.append(temp)
con = sqlite3.connect('ceproject.db')
con.execute('DROP TABLE IF EXISTS europe_Gross_Production_Index_Number ;')
con.execute('''CREATE TABLE europe_Gross_Production_Index_Number(
               country TEXT NOT NULL,
               year INTEGER NOT NULL,
               indexnumber REAL NOT NULL,
               PRIMARY KEY(country, year));''')
con.commit()
for n in newdata:
    con.execute('INSERT INTO europe_Gross_Production_Index_Number VALUES(?, ?, ?);', n)
    con.commit()
con.commit()
con.close();
```

## 1.4  Mortality Attibutable Air Pollution

```python
rawdata3 = pd.read_csv('data/air_quality_health_risk_assessments.csv')
pm2_5data = rawdata3[rawdata3['Air Pollutant'] == 'PM2.5']
pm2_5data = pm2_5data[['Country Or Territory', 'Degree Of Urbanisation', 'Year', 'Premature Deaths', 'Premature Deaths - lower CI', 'Premature Deaths - upper CI', 'Years Of Life Lost', 'Years Of Life Lost - lower CI', 'Years Of Life Lost - upper CI']]
allpm2_5data = pm2_5data[pm2_5data['Degree Of Urbanisation'] == 'All Areas (incl.unclassified)']
urbanpm2_5data = pm2_5data[pm2_5data['Degree Of Urbanisation'] == 'Urban clusters']
ruralpm2_5data = pm2_5data[pm2_5data['Degree Of Urbanisation'] == 'Rural grid cells']
allpm2_5data = allpm2_5data.dropna().reset_index(drop=True).drop(columns='Degree Of Urbanisation')
urbanpm2_5data = urbanpm2_5data.dropna().reset_index(drop=True).drop(columns='Degree Of Urbanisation')
ruralpm2_5data = ruralpm2_5data.dropna().reset_index(drop=True).drop(columns='Degree Of Urbanisation')
allpm2_5datae = []
urbanpm2_5datae = []
ruralpm2_5datae = []
for c in countries_name:
    for a in range(len(allpm2_5data)):
        if c in allpm2_5data['Country Or Territory'][a]:
           allpm2_5datae.append(allpm2_5data.loc[a].tolist())
    for b in range(len(urbanpm2_5data)):
        if c in urbanpm2_5data['Country Or Territory'][b]:
           urbanpm2_5datae.append(urbanpm2_5data.loc[b].tolist())
    for d in range(len(ruralpm2_5data)):
        if c in ruralpm2_5data['Country Or Territory'][d]:
           ruralpm2_5datae.append(ruralpm2_5data.loc[d].tolist())
allpm2_5datae = [[f.item() if isinstance(f, np.int64) else f for f in sublist] for sublist in allpm2_5datae]
urbanpm2_5datae = [[g.item() if isinstance(g, np.int64) else g for g in sublist] for sublist in urbanpm2_5datae]
ruralpm2_5datae = [[h.item() if isinstance(h, np.int64) else h for h in sublist] for sublist in ruralpm2_5datae]
con = sqlite3.connect('ceproject.db')
con.execute('DROP TABLE IF EXISTS pm2_5_lifedeath_all_areas ;')
con.execute('''CREATE TABLE pm2_5_lifedeath_all_areas(
               country TEXT NOT NULL,
               year INTEGER NOT NULL,
               pm2_5_Premature_Deaths INTEGER,
               pm2_5_Premature_Deaths_LCI INTEGER,
               pm2_5_Premature_Deaths_UCI INTEGER,
               pm2_5_Years_Of_Life_Lost INTEGER,
               pm2_5_Years_Of_Life_Lost_LCI INTEGER,
               pm2_5_Years_Of_Life_Lost_UCI INTEGER,
               PRIMARY KEY(country, year));''')
con.commit()
for n in allpm2_5datae:
    con.execute('INSERT INTO pm2_5_lifedeath_all_areas VALUES(?, ?, ?, ?, ?, ?, ?, ?);', n)
    con.commit()
con.execute('DROP TABLE IF EXISTS pm2_5_lifedeath_urban ;')
con.execute('''CREATE TABLE pm2_5_lifedeath_urban(
               country TEXT NOT NULL,
               year INTEGER NOT NULL,
               pm2_5_Premature_Deaths INTEGER,
               pm2_5_Premature_Deaths_LCI INTEGER,
               pm2_5_Premature_Deaths_UCI INTEGER,
               pm2_5_Years_Of_Life_Lost INTEGER,
               pm2_5_Years_Of_Life_Lost_LCI INTEGER,
               pm2_5_Years_Of_Life_Lost_UCI INTEGER,
               PRIMARY KEY(country, year));''')
con.commit()
for m in urbanpm2_5datae:
    con.execute('INSERT INTO pm2_5_lifedeath_urban VALUES(?, ?, ?, ?, ?, ?, ?, ?);', m)
    con.commit()
con.execute('DROP TABLE IF EXISTS pm2_5_lifedeath_rural ;')
con.execute('''CREATE TABLE pm2_5_lifedeath_rural(
               country TEXT NOT NULL,
               year INTEGER NOT NULL,
               pm2_5_Premature_Deaths INTEGER,
               pm2_5_Premature_Deaths_LCI INTEGER,
               pm2_5_Premature_Deaths_UCI INTEGER,
               pm2_5_Years_Of_Life_Lost INTEGER,
               pm2_5_Years_Of_Life_Lost_LCI INTEGER,
               pm2_5_Years_Of_Life_Lost_UCI INTEGER,
               PRIMARY KEY(country, year));''')
con.commit()
for k in ruralpm2_5datae:
    con.execute('INSERT INTO pm2_5_lifedeath_rural VALUES(?, ?, ?, ?, ?, ?, ?, ?);', k)
    con.commit()
con.commit()
con.close();
```

## 1.6 Carbon Emissions 

```python
countries_code = []
for country in countries_name:
    url = f'https://restcountries.com/v3/name/{country}?fullText=true&fields=cca3'
    r = requests.get(url)
    data = r.json()
    countries_code.append(data[0]['cca3'])
datalist = []
for code in countries_code:
    r = requests.get(f'https://www.climatewatchdata.org/api/v1/data/historical_emissions?source_ids[]=160&gas_ids[]=401&sector_ids[]=1859&regions[]={code}')
    data = r.json()
    datalist.append(data)
celist = []
for d in datalist:
    if d['data']:
         for ce in d['data'][0]['emissions']:
             temp = []
             temp.append(d['data'][0]['country'])
             temp.append(int(ce['year']))
             temp.append(float(ce['value']))
             celist.append(temp)
con = sqlite3.connect('ceproject.db')
con.execute('DROP TABLE IF EXISTS all_ghg_emissions_europe ;')
con.execute('''CREATE TABLE all_ghg_emissions_europe(
               country TEXT NOT NULL,
               year INTEGER NOT NULL,
               emission_value REAL NOT NULL,
               PRIMARY KEY(country, year));''')
con.commit()
for n in celist:
    con.execute('INSERT INTO all_ghg_emissions_europe VALUES(?, ?, ?);', n)
    con.commit()
con.commit()
con.close();
   
```

## 2.4 Heat map

```python
con = sqlite3.connect('ceproject.db')
df = pd.read_sql_query('''SELECT *
                          FROM all_ghg_emissions_europe
                          JOIN europe_Gross_Production_Index_Number
                          ON all_ghg_emissions_europe.country = europe_Gross_Production_Index_Number.country
                          AND all_ghg_emissions_europe.year = europe_Gross_Production_Index_Number.year
                          JOIN pm2_5_pollution_all_areas
                          ON all_ghg_emissions_europe.country = pm2_5_pollution_all_areas.country
                          AND all_ghg_emissions_europe.year = pm2_5_pollution_all_areas.year
                          JOIN pm2_5_lifedeath_all_areas
                          ON all_ghg_emissions_europe.country = pm2_5_lifedeath_all_areas.country
                          AND all_ghg_emissions_europe.year = pm2_5_lifedeath_all_areas.year
                          ORDER BY country ASC;''', con)
con.close();
df = df.T.drop_duplicates().T
df = df.drop(columns=['population', 'area', 'pm2_5pollution_weighted', 'pm2_5_Premature_Deaths_LCI', 'pm2_5_Premature_Deaths_UCI', 'pm2_5_Years_Of_Life_Lost_LCI', 'pm2_5_Years_Of_Life_Lost_UCI'])
df = df.astype({"year":'category', "emission_value":'float64', "indexnumber":'float64', "pm2_5pollution":'float64', "pm2_5_Premature_Deaths":'int64', "pm2_5_Years_Of_Life_Lost": 'int64'})

fig, axes = plt.subplots(11, 3, figsize=(20, 50))

for i, country in enumerate(df['country'].unique()):
    row = i // 3
    col = i % 3
    country_data = df[df['country'] == country]
    corr = country_data.corr()
    sns.heatmap(corr, cmap='coolwarm', vmin=-1, vmax=1, annot=True, fmt='.2f', square=True, ax=axes[row, col])
    axes[row, col].set_title(country)

plt.tight_layout()
plt.show()
```
